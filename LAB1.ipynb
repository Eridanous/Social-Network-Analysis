```python
%matplotlib inline
import matplotlib
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

### Α) Δημιουργία και οπτικοποίηση σύνθετων τύπων δικτύου

Κατασκευή και οπτικοποίηση της τοπολογίας REG για βαθμό πλέγματος $d = 4$

```python
d = 4

# για την κατασκευή του REG χρησιμοποιήστε την watts_strogatz_graph από το networkx. Επιλέξτε προσεκτικά τις παραμέτρους. 
REG = nx.watts_strogatz_graph(n,d,0)

# για την οπτικοποίηση, χρησιμοποιήστε την draw_circular από το networkx

nx.draw_circular(REG,node_size=40)
```

Κατασκευή και οπτικοποίηση της τοπολογίας RG(ER) για αριθμό συνδέσεων $M = 750$

```python
M = 750

# για την κατασκευή του RGER χρησιμοποιήστε την gnm_random_graph από το networkx
RGER = nx.gnm_random_graph(n,M)

# για την οπτικοποίηση, χρησιμοποιήστε την draw_circular από το networkx

nx.draw_circular(RGER,node_size=40)
```

Κατασκευή και οπτικοποίηση της τοπολογίας RGG σε περιοχή $L\times L = 1$ (unit cube) και ακτίνα σύνδεσης $R=0.25$

```python
R = 0.25
#L = 1 #de xreaizetai


# για την κατασκευή του RGG, χρησιμοποιήστε την random_geometric_graph από το networkx
RGG= nx.random_geometric_graph(n,R,2)

# για την οπτικοποίηση, χρησιμοποιήστε draw_networkx_edges, draw_networkx_nodes. 
# Για τη θέση των κόμβων στην τετράγωνη περιοχή, χρησιμοποιήστε την get_node_attributes από το networkx.
pos=  nx.get_node_attributes(RGG,'pos')

nx.draw_networkx_edges(RGG,pos)
nx.draw_networkx_nodes(RGG,pos,node_size=40)

```

Κατασκευή και οπτικοποίηση της τοπολογίας SF για βαθμό αρχικού πλέγματος $d=4$

```python
d = 4

# για την κατασκευή του SF, χρησιμοποιήστε την barabasi_albert_graph από το networkx.
SF = nx.barabasi_albert_graph(n,d)

# για την οπτικοποίηση, χρησιμοποιήστε την draw_circular από το networkx.

nx.draw_circular(SF,node_size=40)
```

Κατασκευή και οπτικοποίηση της τοπολογίας SW για βαθμό αρχικού πλέγματος $d=4$ και πιθανότητας ανασύνδεσης $p=0.3$

```python
d = 4
p = 0.3

# για την κατασκευή του SW, χρησιμοποιήστε την watts_strogatz_graph από το networkx.
SW = nx.watts_strogatz_graph(n,d,p)

# για την οπτικοποίηση, χρησιμοποιήστε την draw_circular από το networkx.

nx.draw_circular(SW,node_size=40)
```

<b>Σχολιάστε εδώ τα αποτελέσματα των ερωτημάτων Β1, Β2, Β3.</b>

Στη Regular τοπολογία όλοι οι κόμβοι έχουν ακριβώς 4 συνδέσεις, άρα όλοι έχουν βαθμό 4 και η διασπορά είναι 0.

Στη Randon Graph (ER) τοπολογία, αφού επιλέξαμε να έχουμε $M=750$ συνδέσεις στους $n=100$ κόμβους του δικτύου, το οποίο είναι undirected (άρα κάθε σύνδεσμος "μετράει σαν δύο"), ο μέσος βαθμός του δικτύου είναι $\frac{M}{n} 2 = 15$. 
Όσον αφορά τη διασπορά, από το διάγραμμα κατατανομής βαθμών κόμβων του δικτύου RGER του ερωτήματος Β2, βλέπουμε ότι τώρα οι τιμές "παίζουν" αρκετά γύρω από τη μέση τιμή, γεγονός που δικαιολογεί ότι η τιμή της είναι της ίδιας τάξης μεγέθους με τη μέση τιμή.
Επίσης, επειδή οι συνδέσεις γίνονται με ίση πιθανότητα (και δεν συγκεντρώνει κάποιος κόμβος περισσότερες με preferential attachment, όπως γίνεται στη Scale Free), στη συγκεντρωτική κατανομή βαθμού κόμβου βλέπουμε μικρές μεταβολές στα άκρα και λίγο μεγαλύτερες γύρω από τη μέση τιμή, δηλαδή μια σχετικά γραμμική μεταβολή. 

Στη Randon Geometric Graph τοπολογία, ισχύουν τα ίδια σχεδόν με τη Randon Graph, με τη διαφορά ότι τώρα η πραγματοποίηση των συνδέσεων εξαρτάται από την απόσταση των κόμβων. Οι σύνδέσεις τώρα πραγματοποιούνται αν οι δύο εξεταζόμενοι κόμβοι απέχουν απόσταση μικρότερη της δοθείσης ακτίνας. Επομένως, παίρνουμε μια συμπεριφορά αρκετά παρόμοια με την τοπολογία Randon Graph (ER). Σε αυτή την περίπτωση η διασπορά εξαρτάται σε μεγάλο βαθμό από τις τυχαίες θέσεις των κόμβων.

Στη Scale Free τοπολογία το ποσοστό των κόμβων που συνδέονται με $k$ κόμβους είναι $P(k)\propto k^{-\gamma}$, όπου $2<\gamma<3$ για μεγάλα $k$. Το οποίο πρακτικά σημαίνει ότι μικρό ποσοστό των κόμβων έχουν υψηλό βαθμό και μεγάλο ποσοστό τους έχει μικρό βαθμό. Το γεγονός αυτό φαίνεται στο αντίστοιχο σχήμα του ερωτήματος Β1, αλλά και στα διαγράμματα κατατανομής βαθμών κόμβων και συγκεντρωτικής κατανομής βαθμού κόμβου  του δικτύου SF του ερωτήματος Β2. Επομένως, σε μια τέτοια τοπολογία είναι λογικό να έχουμε πολύ μεγάλη διασπορά γύρω από τη μέση τιμή.

Η Small World τοπολογία προκύπτει από τη Regual Graph με τη διαφορά ότι υπάρχει και πιθανότητα $p$, μια ακμή να μην ενώσει τους "γειτονικούς" κόμβους της Regular αλλά κάποιους άλλους με τυχαίο τρόπο (γι' αυτό αν ορίσουμε $p=0$ θα πάρουμε τη Regular). Η συμπεριφορά αυτή φαίνεται στο διάγραμμα του ερωτήματος Β1. Επομένως, η μέση τιμή παραμένει σταθερή ($d=4$ στην προκειμένη περίπτωση), ενώ η διασπορά είναι πολύ μικρή για τιμή πιθανότητας $p=0.3$.

```python
# χρησιμοποήστε την all_pairs_shortest_path_length του networkx 
def compute_path_lengths(G):
#     return list(nx.all_pairs_shortest_path_length(G))
    lvl1 = list(dict(nx.all_pairs_shortest_path_length(G)).values())
    lvl2 = []
    for i in range(0,n):
        lvl2.append((list(dict(lvl1[i]).values())))
    return lvl2
    
def nozero_matrix(sp):    
    temp = []
    lim = len(sp) #sp = array lim x lim
    for i in range(0,lim):
        for j in range(0,lim):
            temp2 = sp[i][j]
            if temp2 != 0:
                temp.append(temp2)
    return temp           
                

REG_path_lengths = compute_path_lengths(REG)
RGER_path_lengths = compute_path_lengths(RGER)
RGG_path_lengths = compute_path_lengths(RGG)
SF_path_lengths = compute_path_lengths(SF)
SW_path_lengths = compute_path_lengths(SW)

REG_spl = nozero_matrix(REG_path_lengths) #δε θελουμε να συμπεριλάβουμε τα shotest paths από ένα κόμβο στον εαυτό του (που το shortest path είναι μηδέν)
RGER_spl = nozero_matrix(RGER_path_lengths)
RGG_spl = nozero_matrix(RGG_path_lengths)
SF_spl = nozero_matrix(SF_path_lengths)
SW_spl = nozero_matrix(SW_path_lengths)

print("Mean path length REG: %f" %np.mean(REG_spl)) 
print("Variance of path length REG: %f" %np.var(REG_spl))

print("Mean path length RGER: %f" %np.mean(RGER_spl))
print("Variance of path length RGER: %f" %np.var(RGER_spl))

print("Mean path length RGG: %f" %np.mean(RGG_spl))
print("Variance of path length RGG: %f" %np.var(RGG_spl))

print("Mean path length SF: %f" %np.mean(SF_spl))
print("Variance of path length SF: %f" %np.var(SF_spl))

print("Mean path length SW: %f" %np.mean(SW_spl))
print("Variance of path length SW: %f" %np.var(SW_spl))
```

Η Random Geometric Graph τοπολογία καταφέρνει και αυτή να μειώσει το μέσο μήκος μονοπατιού σε σχέση με τη Regular, εξαιτίας της τυχαιότητας που διέπει τις συνδέσεις. Η Scale Free τοπολογία έχει μικρό μέσο μήκος μονοπατιού εξαιτίας των λίγων κόμβων με τις πολλές συνδέσεις. Αυτοί οι κόμβοι είναι λίγοι μεν αλλά συνδέονται με πάρα πολλούς άλλους, λειτουργώντας σαν "κέντρα". Οπότε κάθε κόμβος συνδέται με τους υπόλοιπους μέσω λίγων hops περνώντας κατά πάσα πιθανότητα από αυτούς τους κόμβους "κέντρα" (βλ. αντιστοιχο διάγραμμα στο ερώτημα Α).

Τέλος, η Small World τοπολογία είναι σαν τη Regular αλλά παρουσιάζει μικρότερο μέσο μήκος μονοπατιού εξαιτίας των τυχαίων συνδέσεων - "shortcutrs" (βλ. αντιστοιχο διάγραμμα στο ερώτημα Α). Οι τυχαίες αυτές συνδέσεις, που είναι και το χαρακτηριστικό που την ξεχωρίζει απο τη Regular, επιτρέπουν hops ανάμεσα σε μη "γειτονικούς" κόμβους μειώνοντας αισθητά το μέσο μήκος μονοπατιού της τοπολογίας.

<h4>Αναλυτικός υπολογισμός κεντρικοτήτων</h4>

Degree Centrality

$C_D(i) = \dfrac{d_i}{n-1}$, where $d_i$ the degree of node $i$.

Closeness Centrality

$C_C(i) = \left[ \dfrac{\sum_{j=1}^n}{n-1} d(i,j) \right] ^{-1}$, where $d(i,j)=$ shortest path between nodes $i$ and $j$ (in hops).

Betweenness Centrality

To calculate betweenness centrality, you take every pair of the network and count how many times a node can interrupt the shortest paths (geodesic distance) between the two nodes of the pair. For standardization, I note that the denominator is (n-1)(n-2)/2.

$C_B(i) = \dfrac{count}{\frac{(n-1)(n-2)}{2}}$, where $\frac{(n-1)(n-2)}{2}$ is Maximum possible Betweenness centrality: Centrality of Star or Wheel graph (Freeman)

### Σχολιασμός

Στη <b>Regular</b> τοπολογία, όλοι οι κόμβοι είναι ισοδύναμοι λόγω συμμερτρίας, οπότε σε όλες τις κεντρικότητες όλοι οι κόμβοι έχουν την ίδια τιμή κεντρικότητας.

<b>Degree centrality:</b> Όσο μικρότερος μέσος βαθμός κόμβου, τόσο μικρότερες τιμές degree centrality (βλ. ερώτημα Β). Η διαφορά της Scale Free είναι ότι επειδή υπάρχει προτίμηση (power law) το μεγαλύτερο ποσοστό κόμβων δεν έχει πολλές συνδέσεις, άρα μικρό degree centrality. Αντίθετα, το μικρό ποσοστό των κόμβων που έχουν πάρα πολλές συνδέσεις θα έχει πολύ μεγάλο degree centrality.

<b>Closeness centrality:</b> Όσο μικρότερο mean shoprtest path length, τόσο μεγαλύτερες τιμές closeness centrality (βλ. ερώτημα Δ). Η διαφορά της Scale Free είναι ότι επειδή υπάρχει προτίμηση (power law) το μεγαλύτερο ποσοστό κόμβων δεν έχει πολλές συνδέσεις. Αντίθετα, το μικρό ποσοστό των κόμβων που έχουν πάρα πολλές συνδέσεις, θα απέχουν λίγα hops από τους υπόλοιπους κόμβους και άρα θα έχουν μεγάλη τιμή closeness centrality.

<b>Betweenness centrality:</b>
Στη Regular, επειδή δεν υπάρχουν "shortcuts" οι κόμβοι θα βρίσκονται σε μεγάλο ποσοστό εντός των διαφόρων shortest paths, με αποτέλεσμα να παρατηρείται πολύ υψηλή betweenness centrality. Στις υπόλοιπες τοπολογίες δεν ισχύει αυτό και γι'αυτό παρουσιάζουν αισθητά χαμηλότερη betweenness centrality. Οι Small World, Random Graph (ER) και η Random Graph (Gilbert) παρουσιάζουν αρκετά παρόμοια συμπεριφορά, λόγω της τυχαιότητας των συνδέσεων. Η διαφορά της Scale Free είναι ότι επειδή υπάρχει προτίμηση (power law) το μεγαλύτερο ποσοστό κόμβων δεν έχει πολλές συνδέσεις, άρα θα "παρεμβάλλετεται" σε μικρό αριθμό shortest path και άρα θα έχει μικρή τιμή betweenness centrality. Αντίθετα, το μικρό ποσοστό των κόμβων που έχουν πάρα πολλές συνδέσεις, θα "παρεμβάλλετεται" σε μεγάλο αριθμό shortest path και άρα θα έχει μεγάλη τιμή betweenness centrality.

<b>Eigenvector centrality:</b>
<i>"In graph theory, eigenvector centrality (also called eigencentrality) is a measure of the influence of a node in a network. Relative scores are assigned to all nodes in the network based on the concept that connections to high-scoring nodes contribute more to the score of the node in question than equal connections to low-scoring nodes. A high eigenvector score means that a node is connected to many nodes who themselves have high scores."</i>

Επομένως, παρατηρούμε ότι οι τοπολογίες δεν παρουσιάζουν τεράστιες διαφορές συμπρειφοράς όσον αφορά αυτο το είδος κεντρικότητας. Επίσης, όπως πάντα, ξεχωρίζει το μικρό ποσοστό κόμβων της Scale Free τοπολογίας, οι οποίοι έχουν πάρα πολλές συνδέσεις και άρα είναι (και θα συνδέονται με) κόμβους υψηλού "influence", συγκεντρώνοντας υψηλό "score".

<b>Sharp:</b> Small World (από watts_strogatz_graph), Random Geometric Graph

### Ι) Μελέτη της εξελικτικής μετατροπής δικτύου REG σε δίκτυο SW και RGER

Μεταβάλλοντας την παράμετρο $g_p$ της συνάρτησης που κατασκευάζει τοπολογίες SW, από $0$ έως $1$ με βήμα $0.1$, να βρείτε το μέσο μήκος μονοπατιού και το μέσο συντελεστή ομαδοποίησης. Πώς συμπεριφέρεται το παραγμόμενο δίκτυο για $g_p = 0$, πώς για $g_p = 1$ και πώς για ενδιάμεσες τιμές $g_p \simeq 0.1 - 0.5$;

```python
def spl(G): #shortest path lengths
#     return list(nx.all_pairs_shortest_path_length(G))
    lvl1 = list(dict(nx.all_pairs_shortest_path_length(G)).values())
    lvl2 = []
    for i in range(0,n):
        lvl2.append(list(dict(lvl1[i]).values()))
    return lvl2

d=4

mn_path= [] #mean shortest paths
mn_CC= [] #mean clustering coefficient

lim = np.arange(0, 1.02, 0.1)

for g_p in lim:
    SWtest = nx.watts_strogatz_graph(n,d,g_p)

    mn_path.append(np.mean(spl(SWtest)))
    mn_CC.append(np.mean(list(dict(nx.clustering(SWtest)).values())))
    
# print(mn_CC)

plt.bar(list(lim),mn_path, width=0.05, color='b')
plt.title("SW n = %d, d = %d" %(n,d))
plt.ylabel("Mean shortest path")
plt.xlabel("$g_p$")
plt.show()

plt.bar(list(lim),mn_CC, width=0.05, color='b')
plt.title("SW n = %d, d = %d" %(n,d))
plt.ylabel("Mean clustering coefficient")
plt.xlabel("$g_p$")
plt.show()
```

Σχολιασμός εξελικτικής μετατροπής:

Για $g_p = 0$ έχουμε δίκτυο Regular, αφού η πιθανόητητα να συνδεθούν άλλοι κόμβοι από τους "γειτονικούς" με τυχαίο τρόπο, είναι μηδέν. Άρα το δίκτυο είναι καθορισμένο στη Regular τοπολογία με μεγάλο μέσο μήκος μονοπατιού και συντελεστή ομαδοποίησης ίσο με $0.5$.

Όσο αυξάνεται η πιθανότητα $g_p$, τόσο μειώνεται το μέσο μήκος μονοπατιού και ο συντελεστής ομαδοποίησης. Το δίκτο έχει πλέον τοπολογία Scale Free. Για $g_p=1$ το δίκτυο μεταβαίνει σε Random Graph (ER) τοπολογία, αφού τώρα κάθε ακμή συνδέεται με τυχαίο τρόπο.

### Κ) Εγω-κεντρικότητες (Ego-centralities)
Για τις συνθετικές τοπολογίες του πίνακα 1, να υπολογιστεί η ενδιαμεσική εγω-κεντρικότητα (ego-betweenness centrality) του κάθε κόμβου. Στη συνέχεια να καταταχθούν οι κόμβοι με βάση την ενδιαμεσική εγωκεντρικότητα και την ενδιαμεσική κεντρικότητα και να συγκριθούν/σχολιαστούν οι κατατάξεις ως προς την ομοιότητά τους για κάθε μία τοπολογία.  

```python
%matplotlib inline
import matplotlib
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

n=100


#για την κατασκευή των εγωδικτύων, χρησιμοποιήστε την ego_graph του networkx
#για τον υπολογισμό της ενδιαμεσικής εγωκεντρικότητας θα χρειαστείτε τις adjacency_matrix (networkx), to.dense, ones (numpy)
#linalg.matrix_power, multiply, triu, nonzero του numpy
#κατασκευάστε dataframe από το pandas με γραμμές τους κόμβους και στήλες τις τιμές των υπό εξέταση κεντρικοτήτων
#κατασκευάστε στη συνέχεια pairplot από το seaborn για να εξετάσετε τις συσχετίσεις μεταξύ τους.

d=4

REG = nx.watts_strogatz_graph(n,d,0)

M=750
RGER = nx.gnm_random_graph(n,M)

# p=0.3
# RGGil = nx.gnp_random_graph(n,p)

R=0.25
RGG= nx.random_geometric_graph(n,R,2)

SF = nx.barabasi_albert_graph(n,d)

p=0.3
SW = nx.watts_strogatz_graph(n,d,p)

# calculate ego betweenness centrality
def ego_BC(topology):
    result = []
    result2 = {}
    for v in topology.nodes():
        A = nx.adjacency_matrix(nx.ego_graph(topology,v)).todense()
        k = A.shape[0]
        A_squared = np.linalg.matrix_power(A,2)
        invA = np.ones((k,k))-A
        G =  np.triu(np.multiply(A_squared,invA),1)
        l = G[np.nonzero(G)]
        temp_result = sum([1/x for x in list(l)])
        result.append(temp_result)
        # aDict[key] = value
        result2[v] = temp_result
    return result2

# ego-Betweenness Centrality
REG_ego_BC = ego_BC(REG)
RGER_ego_BC = ego_BC(RGER)
# RGGil_ego_BC = ego_BC(RGGil)
RGG_ego_BC = ego_BC(RGG)
SF_ego_BC = ego_BC(SF)
SW_ego_BC = ego_BC(SW)

# Betweenness Centrality
REG_BC = nx.betweenness_centrality(REG)
RGER_BC = nx.betweenness_centrality(RGER)
# RGGil_BC = nx.betweenness_centrality(RGGil)
RGG_BC = nx.betweenness_centrality(RGG)
SF_BC = nx.betweenness_centrality(SF)
SW_BC = nx.betweenness_centrality(SW)


#Sort
#========================================================================
#### sorted_by_value = sorted(x.items(), key=lambda kv: kv[1])
REG_ego_BC_srt = sorted(REG_ego_BC.items(), key=lambda kv: kv[1])
RGER_ego_BC_srt = sorted(RGER_ego_BC.items(), key=lambda kv: kv[1])
# RGGil_ego_BC_srt = sorted(RGGil_ego_BC.items(), key=lambda kv: kv[1])
RGG_ego_BC_srt = sorted(RGG_ego_BC.items(), key=lambda kv: kv[1])
SF_ego_BC_srt = sorted(SF_ego_BC.items(), key=lambda kv: kv[1])
SW_ego_BC_srt = sorted(SW_ego_BC.items(), key=lambda kv: kv[1])

REG_BC_srt = sorted(REG_BC.items(), key=lambda kv: kv[1])
RGER_BC_srt = sorted(RGER_BC.items(), key=lambda kv: kv[1])
# RGGil_BC_srt = sorted(RGGil_BC.items(), key=lambda kv: kv[1])
RGG_BC_srt = sorted(RGG_BC.items(), key=lambda kv: kv[1])
SF_BC_srt = sorted(SF_BC.items(), key=lambda kv: kv[1])
SW_BC_srt = sorted(SW_BC.items(), key=lambda kv: kv[1])
#========================================================================


REG_egobc = list(REG_ego_BC.values())
REG_bc = list(REG_BC.values())

RGER_egobc = list(RGER_ego_BC.values())
RGER_bc = list(RGER_BC.values())

# RGGil_egobc = list(RGGil_ego_BC.values())
# RGGil_bc = list(RGGil_BC.values())

RGG_egobc = list(RGG_ego_BC.values())
RGG_bc = list(RGG_BC.values())

SF_egobc = list(SF_ego_BC.values())
SF_bc = list(SF_BC.values())

SW_egobc = list(SW_ego_BC.values())
SW_bc = list(SW_BC.values())


#  d = {'col1': [1, 2], 'col2': [3, 4]}
#  df = pd.DataFrame(data=d)

col1 = 'Ego Betw Centrality'
col2 = 'Betwenness Centrality'


d = {col1: REG_egobc, col2: REG_bc}
REG_df = pd.DataFrame(data=d)

d = {col1: RGER_egobc, col2: RGER_bc}
RGER_df = pd.DataFrame(data=d)

# d = {col1: RGGil_egobc, col2: RGGil_bc}
# RGGil_df = pd.DataFrame(data=d)

d = {col1: RGG_egobc, col2: RGG_bc}
RGG_df = pd.DataFrame(data=d)

d = {col1: SF_egobc, col2: SF_bc}
SF_df = pd.DataFrame(data=d)

d = {col1: SW_egobc, col2: SW_bc}
SW_df = pd.DataFrame(data=d)

g=sns.pairplot(REG_df)
g.fig.suptitle("Regular")

g=sns.pairplot(RGER_df)
g.fig.suptitle("Random Graph (ER)")

# g=sns.pairplot(RGGil_df)
# g.fig.suptitle("Random Graph Gilbert")

g=sns.pairplot(RGG_df)
g.fig.suptitle("Random Geometric Graph")

g=sns.pairplot(SF_df)
g.fig.suptitle("Scale Free")

g=sns.pairplot(SW_df)
g.fig.suptitle("Small World")
```

```python
# κατατάσουμε με αύξουσα σειρά τους κόμβους ως προς το betweenness centrality και το ego-betweenness centrality. 
# Θέλουμε να δούμε πόσο καλή προσσέγγιση είναι να υπολογίζουμε το ego-betweenness αντί του betweenness centrality που είναι πιο απαιτητικό υπολογιστικά.
# Δεν υπάρχει κάποια "σωστή" κανονικοποίηση που μπορούμε να κάνουμε στην ego-betweenness έτσι ώστε να ταυτιστεί με την betweenness centrality.
# Το μόνο που μπορούμε να εξετάσουμε ε΄ίναι αν οι κόμβοι έχουν την ίδια κατάταξη ως προς τις δύο κεντρικότητες 
# (και συγκριτικά αν διατηρούνται οι λόγοι,
# πχ αν ο κόμβος i έχει διπλάσια betweenness centrality από τον j να δούμε αν έχει και διπλάσια ego-betweenness centrality)
# (αυτό δε θα το εξάσουμε αλλά θα μπορούσαμε. Τώρα μας ενδιαφέρει απλά η κατάταξη)

message = "Percentage of nodes at the same classification position: %f \n"
message2 = "Top %d highest nodes in rankings (lower to higher):"
# message2 = "Highest %d rankings (lower to higher):"
head = 10


#REG
comp_bc=[]
comp_ego = []
count = 0
for i in range(0,n):
    temp_bc = REG_BC_srt[i][0]
    temp_ego = REG_ego_BC_srt[i][0]
    
    comp_bc.append(temp_bc)
    comp_ego.append(temp_ego)
    
    if temp_bc == temp_ego: 
        count = count + 1

per = count/n
print("Regular")
print(message %per) 

print(message2 %head)
print(comp_bc[n-head:n])
print(comp_ego[n-head:n])
print("\n")


# RGER
comp_bc=[]
comp_ego = []
count = 0
for i in range(0,n):
    temp_bc = RGER_BC_srt[i][0]
    temp_ego = RGER_ego_BC_srt[i][0]
    
    comp_bc.append(temp_bc)
    comp_ego.append(temp_ego)
    
    if temp_bc == temp_ego:
        count = count + 1

per = count/n
print("Random Graph (ER)")
print(message %per) 

print(message2 %head)
print(comp_bc[n-head:n])
print(comp_ego[n-head:n])
print("\n")

#RGG
comp_bc=[]
comp_ego = []
count = 0
for i in range(0,n):
    temp_bc = RGG_BC_srt[i][0]
    temp_ego = RGG_ego_BC_srt[i][0]
    
    comp_bc.append(temp_bc)
    comp_ego.append(temp_ego)
    
    if temp_bc == temp_ego:
        count = count + 1

per = count/n
print("Random Geometric Graph")
print(message %per) 

print(message2 %head)
print(comp_bc[n-head:n])
print(comp_ego[n-head:n])
print("\n")


#SF
comp_bc=[]
comp_ego = []
count = 0
for i in range(0,n):
    temp_bc = SF_BC_srt[i][0]
    temp_ego = SF_ego_BC_srt[i][0]
    
    comp_bc.append(temp_bc)
    comp_ego.append(temp_ego)
    
    if temp_bc == temp_ego:
        count = count + 1

per = count/n
print("Scale Free")
print(message %per) 

print(message2 %head)
print(comp_bc[n-head:n])
print(comp_ego[n-head:n])
print("\n")



#SW
comp_bc=[]
comp_ego = []
count = 0
for i in range(0,n):
    temp_bc = SW_BC_srt[i][0]
    temp_ego = SW_ego_BC_srt[i][0]
    
    comp_bc.append(temp_bc)
    comp_ego.append(temp_ego)
    
    if temp_bc == temp_ego:
        count = count + 1

per = count/n
print("Small World")
print(message %per) 

print(message2 %head)
print(comp_bc[n-head:n])
print(comp_ego[n-head:n])
print("\n")



```
